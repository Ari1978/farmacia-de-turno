```python
from IPython.display import HTML, Audio
import pyttsx3  # Para generar audio a partir de texto

# Función para obtener información de la farmacia (simulada)
def obtener_info_farmacia(farmacia):
    # Simulamos datos de una farmacia con URLs de imágenes y texto
    farmacias = {
        "farmacia a": {
            "imagen": "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQq8fymcAvCPm8Lz8D4-9pUObqtVrTWTJvFU169E9lmzVFf3u1bYvm-xf3TH_KPcTF-2zk&usqp=CAU",  # URL válida de una imagen de farmacia
            "texto": "Farmacia A: Fundada en 1985, Farmacia A ha sido un pilar en la comunidad local. Durante años, ha ofrecido medicamentos de calidad y un servicio al cliente excepcional, y sigue siendo una de las farmacias más confiables de la ciudad.",
            "audio": "Farmacia A está abierta las 24 horas. Cuenta con medicamentos disponibles."
        },
        "farmacia b": {
            "imagen": "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQOkfFQFHD3F7sKs889UbOujBLqXAJvycw2foLj2pTi1DTN8k7vpz8gV3DK4a02Z8awOYk&usqp=CAU",  # Otra URL válida
            "texto": "Farmacia B: Con más de 30 años de experiencia, Farmacia B ha servido a la comunidad ofreciendo no solo productos farmacéuticos, sino también asesoramiento profesional para el cuidado de la salud.",
            "audio": "Farmacia B está abierta de 9 a 18 horas. Algunos medicamentos están fuera de stock."
        }
    }
    return farmacias.get(farmacia.lower(), None)

# Función para generar el contenido según el prompt
def mostrar_contenido(prompt, farmacia):
    # Obtener la información de la farmacia
    info = obtener_info_farmacia(farmacia)
    if not info:
        return "Farmacia no encontrada."
    
    if "imagen" in prompt:
        # Muestra la imagen desde la URL usando HTML para evitar problemas de visualización
        return HTML(f'<img src="{info["imagen"]}" width="500px">')
    elif "texto" in prompt:
        # Muestra el texto con la historia de la farmacia
        return info["texto"]
    elif "audio" in prompt:
        # Convertir el texto en audio usando pyttsx3
        engine = pyttsx3.init()
        engine.save_to_file(info["audio"], "farmacia_info.mp3")  # Guardar el archivo de audio
        engine.runAndWait()
        return Audio("farmacia_info.mp3")  # Reproduce el archivo de audio
    else:
        return "No se ha encontrado contenido válido en el prompt."

# Función para pedir el input del usuario
def obtener_input_usuario():
    # Solicitar al usuario que ingrese el tipo de contenido que quiere ver
    prompt = input("¿Qué tipo de contenido deseas ver? (imagen/texto/audio): ").lower()
    
    # Solicitar al usuario que ingrese el nombre de la farmacia
    farmacia = input("Ingresa el nombre de la farmacia (Farmacia A/Farmacia B): ").strip().lower()

    # Mostrar el contenido basado en la respuesta del usuario
    return mostrar_contenido(prompt, farmacia)

# Ejecutar la función interactiva
obtener_input_usuario()

```
